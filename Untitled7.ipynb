```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#load the dataset
claimants = pd.read_csv("claimants.csv")
```

```python
claimants.head()
```

```python
#dropping case number columns as it is not required
claimants.drop(["CASENUM"],inplace=True,axis=1)
```

```python
claimants
```

```python
#no.of rows and columns
claimants.shape
```

```python
#removing na values in dataset
claimants = claimants.dropna()
claimants.shape
```

```python
#dividing the data in to x and y variables because logistic regression expect x and y variables in different dataframes
x=claimants.iloc[:,1:]
y=claimants.iloc[:,0]
```

```python
#creating object for logisticregression and we are calling fit function and supplying the data
classifier = LogisticRegression()
classifier.fit(x,y)
```

```python
#once model is built, then apply model on the x dataset, predict for x dataset
y_pred = classifier.predict(x)
```

```python
y_pred
```

```python
#we are creating dataframe by combining outcomes with the actual data
y_pred_df = pd.DataFrame({'actual':y, 'predicted_prob' : classifier.predict(x)})
```

```python
y_pred_df
```

```python
#confusion matrix for accuracy, we are supplying actual data y and predicted values from model
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
```

```python
confusion_matrix
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
#ROC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
#here we are going to compute the probalilities of predicted values for the x and we are get probability for 1st coulmn
fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba (x) [:,1])
# we will get fpr,tpr,threshold values then we r gng to use them in roc plot and we will get area under curve(auc)
auc = roc_auc_score(y, y_pred)
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color = 'red', label = 'logit model(area = %0.2f)'%auc)
plt.plot([0,1], [0,1],'k--')# x and y axis ranges and k-- means the diagonal dotted line in graph
plt.xlabel('false positive rate or [1- true negative rate]')
plt.ylabel('true positive rate')
```
